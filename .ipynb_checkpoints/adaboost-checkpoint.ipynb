```python
## ADABoosting dependencies
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
testData = pd.read_csv(r'C:\Users\asawh\Documents\Code\exoTest.csv')
trainData = pd.read_csv(r'C:\Users\asawh\Documents\Code\exoTrain.csv')
```

```python
testData.head()
```

```python
trainData.head()
```

```python
testData['LABEL'] -= 1
trainData['LABEL'] -= 1
trainData.head()
```

```python
X_train = trainData.iloc[:,1:].copy()
Y_train = trainData['LABEL']
```

```python
X_train.head()
```

```python
classifier = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1),
    n_estimators=200
)

classifier.fit(X_train, Y_train)
```

```python
train_predictions = classifier.predict(X_train)
```

```python
result = confusion_matrix(Y_train, train_predictions)
```

```python
result
```

```python
from sklearn.metrics import accuracy_score
print ('Accuracy = {:0.2f}%'.format(100.0 * accuracy_score(Y_train, train_predictions)))
```

```python
X_test = testData.iloc[:,1:].copy()
Y_test = testData['LABEL']
```

```python
test_predictions = classifier.predict(X_test)
print ('Accuracy = {:0.2f}%'.format(100.0 * accuracy_score(Y_test, test_predictions)))
```

```python
confusion_matrix(Y_test, test_predictions)
```
